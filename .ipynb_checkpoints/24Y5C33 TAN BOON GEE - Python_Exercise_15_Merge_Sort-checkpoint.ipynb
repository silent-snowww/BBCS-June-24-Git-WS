```python
def mergesort(arr, low, high): #this function does the splitting
    if low == high:
        return arr
    else: 
        new = (high + low) // 2 
        #newlist1 = arr[low:new+1]
        #newlist2 = arr[new+1:high+1]
        #print(newlist1, newlist2)
        #print(newlist1[new], newlist2[new])


        arr_left = mergesort(arr, low, new) 
        arr_right = mergesort(arr, new+1, high)
        print(arr_left, arr_right)
        return merge(arr_left, arr_right)
    
    

def merge(left, right): #this function does the merging of the 2 sorted list into 1 bigger sorted list
    if left > right: 
        return [right, left]
    elif right > left:
        return [left, right]


#main
alist = [76, 47, 45, 36, 24, 17, 10, 5]
sorted_list = mergesort(alist, 0, len(alist) - 1)
print(sorted_list)
```
