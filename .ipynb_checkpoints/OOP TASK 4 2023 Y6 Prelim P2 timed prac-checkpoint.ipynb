```python
#Task 4.1 


class Person:
    def __init__(self, pnric, pname, pdob, pvacstatus):
        self.__nric = pnric            #string
        self.__name = pname            #string
        self.__date_of_birth = pdob    #string YYYY-MM-DD
        self.__vaccination_status = pvacstatus   #Boolean
    
    #getters
    def get_nric(self):
        return self.__nric
    
    def get_name(self):
        return self.__name
    
    def get_date_of_birth(self):
        return self.__date_of_birth
    
    def get_vaccination_status(self):
        return self.__vaccination_status
    
    #setters
    def set_nric(self, newnric):
        self.__nric = newnric
        
    def set_name(self, newname):
        self.__name = newname 
        
    def set_date_of_birth(self, newdob):
        self.__date_of_birth = newdob
        
    def set_vaccination_status(self, newstatus):
        self.__vaccination_status = newstatus 
    
    #other methods
    def get_age(self):
        currentyear = 2024 
        birthyear = int(self.__date_of_birth[:4])
        age = currentyear - birthyear 
        return age
    
    def print_details(self):
        print("NRIC:", self.__nric)
        print("Name:", self.__name)
        print("Date of Birth:", self.__date_of_birth)
        print("Vaccination_status:", self.__vaccination_status)
        print()

#testing 
person1 = Person("T0745312Z", "Tom", "2010-03-03", False)
person1.print_details()
print(person1.get_age())
```

```python
#Task 4.2 

import csv
infile = open("unvaccinated.csv", "r")
lines = infile.readlines()
uv_people = []

for i in range(len(lines)):
    lines[i] = lines[i].strip("\n")
    lines[i] = lines[i].split(",")

#print(lines)
    
    temp = Person(lines[i][0], lines[i][1], lines[i][2], lines[i][3])
    uv_people.append(temp)
    
#print(uv_people)
#print()
for people in uv_people:
    people.print_details()

infile.close()
```

```python
#Task 4.3 

#bubble sort
def sorting_by_age(arr):
    totalpass = len(arr) - 1
    currentpass = 1

    while currentpass <= totalpass:
        for i in range(1, len(arr)):
            if arr[i].get_age() > arr[i-1].get_age():
                temp = arr[i]
                arr[i] = arr[i-1]
                arr[i-1] = temp
                
        currentpass = currentpass + 1
                
    return arr

#print(sorting_by_age([5, 4, 7, 3, 6]))
info = sorting_by_age(uv_people)
for people in info:
    people.print_details()
```
