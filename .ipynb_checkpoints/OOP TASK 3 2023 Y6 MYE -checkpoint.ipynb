```python
#Task 3.1 
#include getters and setters for all methods!!

class Student: 

    def __init__(self, sid, full, gen, dob, subjcombi):
        self.__studentid = sid 
        self.__full_name = full
        self.__gender = gen
        self.__date_of_birth = dob
        self.__subject_combination = subjcombi 
        
    def show_details(self):
        print(f"Student ID: {self.__studentid} \nFull name: {self.__full_name} \nGender: {self.__gender} \nDate of birth: {self.__date_of_birth} \nSubject combination: ")
        
        for i in range(len(self.__subject_combination)):
            print (f"Subject {i+1}: {self.__subject_combination[i]}")
    
    #accessors / getters 
    def get_studentid(self):
        return self.__studentid
    
    def get_full_name(self):
        return self.__full_name
    
    def get_gender(self):
        return self.__gender 
    
    def get_date_of_birth(self):
        return self.__date_of_birth
    
    def get_subject_combination(self):
        return self.__subject_combination
    
    # mutators / setters
    
    def set_studentid(self, newsid):
        self.__studentid = newsid
    
    def set_full_name(self, newfull):
        self.__full_name = newfull
    
    def set_gender(self, newgen):
        self.__gender = newgen
    
    def set_date_of_birth(self, newdob):
        self.__date_of_birth = newdob
    
    def set_subject_combination(self, newsubjs):
        self.__subject_combination = newsubjs
    
```

```python
#Task 3.3 

def binary_search_r(arr, search_item, first, last):
    midpoint = (first + last) // 2   # midpoint is index of the midpoint

    if last < first:   # student not in list
        return -1 
    
    if search_item == arr[midpoint]:
        return midpoint 
    
    else: 
        if search_item < arr[midpoint]: 
            last = midpoint - 1
        
        if search_item > arr[midpoint]:
            first = midpoint + 1
    
        return binary_search_r(arr, search_item, first, last)
    
    
def main():
    idlist = [] #not recommended! efficiency will be O(N) instead of O(log N): will be iterating through the list
    for i in range(len(students)):
        ids = students[i].get_studentid()
        idlist.append(int(ids))
    #print(idlist)

    repeat = True
    
    while repeat == True: 
        idnum = int(input("Please enter a student ID. "))

        search = binary_search_r(idlist, idnum, 0, len(students)-1)

        if search == -1:
            print("Student Not Found")
            userchoice = input("Enter 'Yes' if you would like to search again. ")
            if userchoice == "Yes":
                repeat = True
            else:
                repeat = False
        
        elif search != -1:
            repeat = False
            students[search].show_details()
    
        
main()
```

```python
#Task 3.3 corrections

def binary_search_recursive(arr, search_item, first, last):
    midpoint = (first + last) // 2
    
    if first > last:
        return -1 
    
    if search_item == arr[midpoint].get_studentid():
        return midpoint
    
    else:
        if search_item < arr[midpoint].get_studentid():
            last = midpoint - 1
        
        if search_item > arr[midpoint].get_studentid(): 
            first = midpoint + 1
        
        return binary_search_recursive(arr, search_item, first, last)

def newmain():
    answer = "Yes"
    while answer == "Yes": 
        idnum = int(input("Please enter a student ID. "))

        search = binary_search_recursive(students, idnum, 0, len(students)-1)

        if search == -1:
            print("Student Not Found")
            answer = input("Enter 'Yes' if you would like to search again. ")
        
        elif search != -1:
            answer = "No"
            students[search].show_details()

newmain()
```

```python
#validation: validation is not required for this question (question did not request!) 
def validation(sid, full, gen, dob, subjcombi):
    validflag = True 
    while validflag == True:
        if sid.isdigit() == False:
            print("sid error")
            validflag = False

        if type(full) != str: 
            print("full error")
            validflag = False

        if gen != "Male" and gen != "Female":
            print("gen error")
            validflag = False

        if type(dob) != str or len(dob) != 10 or dob[4] != "-" or dob[7] != "-" or not dob[:4].isdigit() \
        or not dob[5:7].isdigit() or not dob[8:].isdigit():
                print("dob error")
                validflag = False

        if type(subjcombi) != list: 
            print("subjcombi error")
            validflag = False
        
        return validflag
        
print(validation("1", "Tom Goh", "Male", "2010-10-30", ["Math", "English"] ))
```
