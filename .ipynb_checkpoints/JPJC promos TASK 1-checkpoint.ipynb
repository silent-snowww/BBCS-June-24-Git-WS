```python
#Task 1.1 

infile = open("Task1_1.txt", "r") 
lines = infile.readlines() 
#print(lines)

date = []
temps = []
averages = []
highest = []
lowest = []

for i in range(len(lines)):   #use this instead of for line in lines! for line in lines is an external variable, doesn't directly edit the original lines list.
    lines[i] = lines[i].strip("\n")
    lines[i] = lines[i].split()
    date.append(lines[i][0])
    temps.append(lines[i][1:])

#print(lines)
#print()
#print(date)
#print()
#print(temps)
#print()

def calculate_average(arr):
    total = 0
    for i in range(len(arr)):
        total = total + float(arr[i])
    average = round((total / len(arr)), 1)

    return average

#print(calculate_average(temps[0]))

def bubble_sort_i(arr):      #sort in ascending order
    totalpass = len(arr)-1
    currentpass = 1
    
    while currentpass <= totalpass: 
        for i in range(1, len(arr)):
            if arr[i-1] > arr[i]:
                temp = arr[i-1]
                arr[i-1] = arr[i]
                arr[i] = temp
        currentpass = currentpass + 1
    
    return arr

def find_max(arr):
    arr = bubble_sort_i(arr)
    return arr[-1]

def find_min(arr):
    arr = bubble_sort_i(arr)
    return arr[0]

    
for j in range(len(temps)):
    #print(bubble_sort_i(temps[j]))
    temp_max = find_max(temps[j])
    highest.append(temp_max)
    temp_min = find_min(temps[j])
    lowest.append(temp_min)
    temp_avg = calculate_average(temps[j])
    averages.append(temp_avg)

#print(highest)
#print(lowest)
#print(averages)

print("Date         Average        Highest        Lowest   ")
for i in range(len(lines)):
    print(f"{date[i]}   {averages[i]}           {highest[i]}           {lowest[i]}")


infile.close()
```

```python
#Task 1.2 

import random 

temp_list = []
feb_max = []
feb_min = []
feb_avg = []
feb_days = 28

def generate_temp():
    random_temp_list = []
    for i in range(5):
        random_temp = random.randint(15,25)
        random_temp_list.append(random_temp)
        
    return random_temp_list


for i in range(feb_days):
    rand_temp = generate_temp()
    #print(rand_temp)
    temp_list.append(rand_temp)

    
for j in range(len(temp_list)):
    temp_max = find_max(temp_list[j])
    feb_max.append(temp_max)
    temp_min = find_min(temp_list[j])
    feb_min.append(temp_min)
    temp_avg = calculate_average(temp_list[j])
    feb_avg.append(temp_avg)

#print(feb_max)
#print()
#print(feb_min)
#print()
#print(feb_avg)

feb_dates = []
for i in range(feb_days):
    one_date = str(i+1)
    if len(str(i+1)) == 1:
        one_date = "0" + str(i+1)
        
    feb_dates.append(f"2023-02-{one_date}")
    
    
#print(temp_list)
#print()
    
#print(feb_dates)
#print()

print("Date         Average        Highest      Lowest   ")
for j in range(len(feb_dates)):
    print(f"{feb_dates[j]}   {feb_avg[j]}           {feb_max[j]}           {feb_min[j]}")
    

outfile = open("simulated_data.txt", "w")
outfile.write("Date         Average        Highest      Lowest   \n")
for j in range(len(feb_dates)):
    data = (f"{feb_dates[j]}   {feb_avg[j]}           {feb_max[j]}           {feb_min[j]} \n")
    outfile.write(data)
outfile.close()
```
