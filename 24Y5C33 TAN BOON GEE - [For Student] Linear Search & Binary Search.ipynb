```python
def binary_search_recursive(arr, search_item, first, last): #can have >1 base case
    midpoint = (first + last) // 2
    
    if last < first:
        return -1 
    
    if search_item == arr[midpoint]:
        return midpoint 
    
    else:
        if search_item < arr[midpoint]:
            last = midpoint - 1
            
        if search_item > arr[midpoint]:
            first = midpoint + 1
        
        return binary_search_recursive(arr, search_item, first, last)


# main
list_of_numbers = [1, 2, 3, 4, 5, 6, 7, 8] # sorted
print(binary_search_recursive(list_of_numbers, 3, 0, len(list_of_numbers)-1)) # found
print(binary_search_recursive(list_of_numbers, 12, 0, len(list_of_numbers)-1)) # not found
```

```python
def binary_search_iterative(arr, search_item, first, last): #for duplicates: linear search method
    found = False
    found_index = -1

    while first <= last and found == False: 
        midpoint = (last + first) // 2    

        if search_item == arr[midpoint]:
            while arr[midpoint] == arr[midpoint-1]: 
                found_index = midpoint - 1
                midpoint = midpoint - 1
            found = True 
            
        elif search_item > arr[midpoint]: 
            first = midpoint+1

        else: #search_item < arr[midpoint]
            last = midpoint-1
        
            
    return found_index 

# main
list_of_numbers = [1, 2, 5, 5, 5, 5, 6, 7, 8] # sorted
print(binary_search_iterative(list_of_numbers, 5, 0, len(list_of_numbers)-1)) # found
#print(binary_search_iterative(list_of_numbers, 12, 0, len(list_of_numbers)-1)) # not found
```
