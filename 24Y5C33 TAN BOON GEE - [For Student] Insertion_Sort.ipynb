```python
def insertion_sort(A):
    
    for i in range(1,len(A)):
        temp = A[i]
        previous = i - 1 #compare with element in front
        while previous >= 0 and temp < A[previous]: #when previous = 0: reached start of array  
                                                    #& checks if the element on the right is smaller than
                                                    #element on left
            A[previous+1] = A[previous]  #duplicate bigger element
            previous = previous - 1 #to check with subsequent elements in front if bigger or smaller  
            #print(A)
        A[previous + 1] = temp  #when previous = -1, put temp at the start of the array 
                                #or if previous != -1, means loop ends because temp > A[previous]:
                                #then put them on the right of A[previous]
    return A

#nested loop
    
def main():
    x = [5,2,4,6,1,3]
    insertion_sort(x)
    print(x)

main()

#bubble sort: only compares with adjacent element 
#insertion sort: compares with each element before inserting at correct place
```

```python
def insertion_sortr(Arr, n):
    if n > 0:
        insertion_sortr(Arr, n-1)
        itemToInsert = Arr[n-1]
        i = n - 2
        while i >= 0 and Arr[i] > itemToInsert:
            Arr[i+1] = Arr[i]
            i = i - 1
        Arr[i+1] = itemToInsert
    else: #n
        return 
    
    #return empty because pass by reference: any changes are made directly to original array
    
        
    
    
def main():
    x = [5,2,4,6,1,3]
    insertion_sortr(x, len(x))
    print(x)

main()
```
