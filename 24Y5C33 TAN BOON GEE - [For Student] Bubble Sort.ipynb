```python
#Bubble Sort Version 1 (without flag)
def bubble_sort_1(arr):
    temp = []
    totalpass = len(arr) - 1
    currentpass = 1
    
    while currentpass <= totalpass:
        i = 0  #reset i to 0 for new pass
        while i <= (totalpass - currentpass):
            if arr[i] > arr[i+1]:
                temp = arr[i+1]
                arr[i+1] = arr[i]
                arr[i] = temp
            i = i + 1
        currentpass = currentpass + 1
    

#need to use nested while loop: now only loop once

def main():
    x = [23,5,8,13,12]
    bubble_sort_1(x)
    print(x)

main()
```

```python
#Bubble Sort Version 2 (with flag)
def bubble_sort_1(arr):
    temp = []
    totalpass = len(arr) - 1
    currentpass = 1
    count = 0
    
    while currentpass <= totalpass:
        validflag = False
        i = 0  
        while i <= (totalpass - currentpass):
            if arr[i] > arr[i+1]:
                temp = arr[i+1]
                arr[i+1] = arr[i]
                arr[i] = temp
                validflag = True
            
            i = i + 1
            count = count + 1
        if validflag == False:
            break
                
        currentpass = currentpass + 1
    
    print(count)


def main():
    x = [23,5,8,13,12]
    bubble_sort_1(x)
    print(x)

main()
```
