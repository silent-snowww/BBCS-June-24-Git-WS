```python
#Task 3.1 

import datetime 

class Student:
    def __init__(self, sname, sgen, sdob):
        self.__name = sname 
        self.__gender = sgen
        self.__date_of_birth = sdob    #YYYY-MM-DD
        
    #getters
    def get_name(self):
        return self.__name

    def get_gender(self):
        return self.__gender
    
    def get_date_of_birth(self):
        return self.__date_of_birth
    
    #setters 
    def set_name(self, newname):
        self.__name = newname 
        
    def set_gender(self, newgender):
        self.__gender = newgender 
        
    def set_date_of_birth(self, newdob):
        self.__date_of_birth = newdob
        
    #other methods
    def get_age(self):
        currentyear = datetime.datetime.now().year
        birthyear = int(self.__date_of_birth[:4])
        age = currentyear - birthyear
        return age
    
#testing data
#student1 = Student("tom", "male", "2010-04-12")
#print(student1.get_age())
```

```python
#Task 3.2

class JuniorHigh(Student):
    def __init__(self, jname, jgen, jdob):
        super().__init__(jname, jgen, jdob)
        
    def get_zoom_name(self):
        studentname = super().get_name()
        
        #removing punctuation
        if studentname.isalpha() == False: 
            newname = ""
            for i in range(len(studentname)):
                if studentname[i].isalpha() == True and studentname[i] != " ":
                    newname = newname + studentname[i]
        
        #removing blank spaces
        elif studentname.isalpha() == True:
            newname = studentname.replace(" ", "")
    
        
        studentage = super().get_age()
        zoom_name = (f"[JH]{newname}{studentage}")
        return zoom_name


class SeniorHigh(Student):
    def __init__(self, hname, hgen, hdob, hhouse):
        super().__init__(hname, hgen, hdob)
        self.__house = hhouse
        
    def get_house(self):
        return self.__house
    
    def set_house(self, newhouse):
        self.__house = newhouse
    
    def get_zoom_name(self):
        studentname = super().get_name()
        studentname = studentname.replace(" ", "")
        studentage = super().get_age()
        zoom_name = (f"[SH]{studentname}{studentage}")
        return zoom_name
        
#testing 
#student2 = JuniorHigh("Jane Tan", "female", "2010-09-23")
#student3 = SeniorHigh("Bob Lim Ha", "male", "2014-05-23", "drakon")
#student4 = JuniorHigh("Lim-Hi-Ha So", "male", "2004-06-07")
#print(student2.get_zoom_name())
#print(student3.get_zoom_name())
#print(student4.get_zoom_name())
```

```python
#Task 3.3 

import csv 

infile = open("STUDENTS.csv", "r")
lines = infile.readlines()
#print(lines)

student_list = []

for i in range(len(lines)):
    lines[i] = lines[i].strip("\n")
    lines[i] = lines[i].split(",")
    
#print(lines)

    #Senior High students have 4 sets of data
    if len(lines[i]) == 4:
        tempstudent = SeniorHigh(lines[i][0], lines[i][1], lines[i][2], lines[i][3])
        student_list.append(tempstudent)
    
    #Junior High students have 3 sets of data
    elif len(lines[i]) == 3:
        tempstudent = JuniorHigh(lines[i][0], lines[i][1], lines[i][2])
        student_list.append(tempstudent)


#print(student_list)

student_list.sort(key=lambda x: x.get_age())
for student in student_list:
    print(student.get_zoom_name())
    

infile.close()
```
