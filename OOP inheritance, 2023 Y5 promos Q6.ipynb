```python
#Task 6.1 

import datetime 

class Vehicle: 
    #constructor
    def __init__(self, vdor, vbrand, vmodel, vcap):
        self.__date_of_registration = vdor
        self.__brand = vbrand
        self.__model = vmodel
        self.__capacity = vcap
    
    #getters
    def get_date_of_registration(self):
        return self.__date_of_registration 
    
    def get_brand(self):
        return self.__brand 
    
    def get_model(self):
        return self.__model
    
    def get_capacity(self):
        return self.__capacity 
    
    #setters
    def set_date_of_registration(self, new_dor):
        self.__date_of_registration = new_dor
    
    def set_brand(self, new_brand):
        self.__brand = new_brand 
        
    def set_model(self, new_model):
        self.__model = new_model
        
    def set_capacity(self, new_capacity):
        self.__capacity = new_capacity 
    
    #other methods
    def display(self): 
        print("Date of registration:", self.__date_of_registration)
        print("Brand:", self.__brand)
        print("Model:", self.__model)
        print("Capacity:", self.__capacity)
        
    def get_num_years_remaining(self):
        date_of_reg = self.__date_of_registration
        year_of_reg = int(date_of_reg[:4])
        current_year = datetime.datetime.now().year
        vage = current_year - year_of_reg
        return (10 - vage)
```

```python
#Task 6.2 

class Car(Vehicle):
    def __init__(self, cdor, cbrand, cmodel, ccap, ctype):
        super().__init__(cdor, cbrand, cmodel, ccap)
        self.__type = ctype 
    
    def get_type(self):
        return self.__type
    
    def set_type(self, newtype):
        self.__type = newtype
        
    def display(self):
        print("----------")
        print("Car")
        print("----------")
        super().display()
        print("Type:", self.__type)
        print()

class Bus(Vehicle):
    def __init__(self, bdor, bbrand, bmodel, bcap, bftype, bnumdoors):
        super().__init__(bdor, bbrand, bmodel, bcap)
        self.__fuel_type = bftype 
        self.__num_of_passenger_doors = bnumdoors
    
    def get_fuel_type(self):
        return self.__fuel_type
    
    def get_num_of_passenger_doors(self):
        return self.__num_of_passenger_doors
    
    def set_fuel_type(self, newftype):
        self.__fuel_type = newftype 
        
    def set_num_of_passenger_doors(self, newnum):
        self.__num_of_passenger_doors = newnum 
        
    def display(self):
        print("----------")
        print("Bus")
        print("----------")
        super().display()
        print("Fuel type:", self.__fuel_type)
        print("Number of doors:", self.__num_of_passenger_doors)
        print()

#main
car1 = Car("2017-10-25", "Honda", "Civic", 5, "Sedan")
car1.display()
print()

bus1 = Bus("2022-01-12", "MAN", "A22", 91, "Electric", 2)
bus1.display()
```

```python
#Task 6.3

import csv 

def load(filename):
    vlist = []
    infile = open(filename, "r")
    lines = infile.readlines()
    #print(lines)
    for i in range(len(lines)):
        lines[i] = lines[i].strip("\n")
        lines[i] = lines[i].split(",")

    #print(lines)

        if len(lines[i]) == 5:
            newcar = Car(lines[i][0], lines[i][1], lines[i][2], int(lines[i][3]), lines[i][4])
            vlist.append(newcar)
            
        elif len(lines[i]) == 6:
            newbus = Bus(lines[i][0], lines[i][1], lines[i][2], int(lines[i][3]), lines[i][4], int(lines[i][5]))
            vlist.append(newbus)
            
    #print(vlist)
    
    return vlist
    infile.close()
    

velist = load("FLEET.csv")
for j in range(len(velist)):
        velist[j].display()
```

```python
#Task 6.4 

def quicksort(alist, attribute):    
    
    if len(alist) <= 1:
            return alist 
    
    if attribute == "years":
        pivot = alist[-1]
        less = []
        more = []
        
        for i in range(len(alist)-1): 
            if alist[i].get_num_years_remaining() >= pivot.get_num_years_remaining():
                more.append(alist[i])
            elif alist[i].get_num_years_remaining() < pivot.get_num_years_remaining(): 
                less.append(alist[i])

            
    elif attribute == "brand":
        pivot = alist[-1]
        less = []
        more = []
        
        for j in range(len(alist)-1):
            if alist[j].get_brand() >= pivot.get_brand():
                more.append(alist[j])
            elif alist[j].get_brand() < pivot.get_brand():
                less.append(alist[j])
    
    less = quicksort(less, attribute)
    more = quicksort(more, attribute)

    return less + [pivot] + more

#main
velist = load("FLEET.csv")
ysorted = quicksort(velist, "years")
print("Sorted by number of years remaining: ")   #should be in ascending number of years remaining!!
print()

for i in range(len(ysorted)):
    ysorted[i].display()

print("=======================")
print("Sorted by brand name: ")
print("=======================")

print()
    
bsorted = quicksort(velist, "brand")
for j in range(len(bsorted)):
    bsorted[j].display()
```
